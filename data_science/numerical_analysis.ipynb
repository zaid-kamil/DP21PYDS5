```python
s1 = [23,45,67,78,90]
s2 = [56,78,45,67,89]
s3 = [67,78,89,90,34]
cl1 = [s1,s2,s3]
```

```python
cl1
```

```python
for std in cl1:
    avg = sum(std)/len(std)
    print(f"{avg} marks")
```

```python
eng = []
for std in cl1:
    eng.append(std[0])
print(sum(eng)/len(eng),'eng average')
```
